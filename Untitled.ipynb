```python
from blackscholes import Option
from portfolio import Position, MultiLeg
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
DAY_COUNT = 365

spot = 520.50
rf30 = 0.05355
rf365 = 0.05
dy = 0.0138

long_put_k = 500
long_put_vol = 0.14
long_put_dte = 25

short_put_k = 500
short_put_vol = 0.172
short_put_dte = 361



long_put = Option('put', dy, rf365, long_put_vol, spot, long_put_dte/DAY_COUNT, long_put_k)
short_put = Option('put', dy, rf365, short_put_vol, spot, short_put_dte/DAY_COUNT, short_put_k)

synth_stock = Option('call', 0, 0, 0.01, spot, 1, 0.00001)
```

```python
print(f"short: {short_put.value()}")
print(f"long: {long_put.value()}")
```

```python
long_put_pos = Position('long_put',10, long_put)
short_put_pos = Position('short_put', -2, short_put)

synth_stock_pos = Position('deep_itm_put',2,synth_stock)

book = [long_put_pos, short_put_pos, synth_stock_pos]

multileg = MultiLeg(book, spot=spot)

```

```python
short_put_pos.value() + long_put_pos.value()
```

```python
multileg.value()
```

```python
multileg.pos_prices()
```

```python
spot_shocks = range(-50,55,5)
spot_levels = []
for s in spot_shocks:
    spot_levels.append(spot + s)

vol_levels = [0.17, 0.18, 0.19, 0.2, 0.21, 0.22, 0.23]
vol_shocks = np.array(range(-10,11,1))/100

days_fwd = 5

shock_values_spot = multileg.value_by_spot(spot_levels=spot_levels, days_fwd=days_fwd, at_expiration=False)
shock_values_spot = pd.Series(data=shock_values_spot, index=spot_levels, name=f"Value by Spot\n{days_fwd} Days Forward")
shock_values_spot_vol = multileg.value_grid_spot_vol(spot_levels=spot_levels, vol_levels=vol_levels, days_fwd=days_fwd)
shock_values_spot_vol = pd.DataFrame(data=shock_values_spot_vol, index=vol_levels, columns=spot_levels)
shock_values_spot_vol_shocks = multileg.value_grid_spot_vol_shocks(spot_shocks=spot_shocks, vol_shocks=vol_shocks, days_fwd=days_fwd)
shock_values_spot_vol_shocks = pd.DataFrame(data=shock_values_spot_vol_shocks, index=vol_shocks, columns=spot_shocks)

shock_pnl_spot = shock_values_spot - multileg.value()
shock_pnl_spot_vol = shock_values_spot_vol - multileg.value()
shock_pnl_spot_vol_shocks = shock_values_spot_vol_shocks - multileg.value()
```
